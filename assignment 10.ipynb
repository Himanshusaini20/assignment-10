### Q1 Explain why we have to use the exception class while creating a custom exception.

```python
""" Built-in exceptions offer information about python related problems and custom 
exception will add information about project related problems. That way you can design 
your code in a way that combines python code with the language of the project."""
```

```python
import inspect as ipt
def tree_class (cls,ind=0):
    print ('-'* ind ,cls.__name__)
    for K in cls.__subclasses__():
        tree_class(K,ind+3)
print ("inbuilt exception is :")
ipt.getclasstree(ipt.getmro(Exception))
tree_class(Exception)
```

### Q4 Why lookup error class is used? Explain with an example keyerror and indexerror.

```python
# The base class for the exceptions that are raised when a key or index used on a 
# mapping or a sequence is invalid.
```

```python
try:
    d={"key": "him",1:[2,3,4,5]}
    print(d["key2"])
except KeyError as e :
    print(e)
```

```python
try:
    l=[1,2,3,4,5]
    print (l[6])
except IndexError as e:
    print (e)
```
